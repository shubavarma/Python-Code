```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\Subhadra\Documents\Python Scripts\Iris.csv",sep=",")
```

```python
datac=data.copy()
```

```python
datac.head()
```

```python
datac.drop(["Id"],axis=1,inplace=True)
```

```python
datac.head()
```

```python
datac["Species"].unique()
```

```python
# split dataset in to 3 datasets depending on species
setosa=datac.loc[datac["Species"]=='Iris-setosa']
versicolor=datac.loc[datac["Species"]=='Iris-versicolor']
virginica=datac.loc[datac["Species"]=='Iris-virginica']
# For 3 datasets if we want to plot using 4 feature variables we will get 6 different combinations
```

```python
# Plot SepalLengh vs SepalWidth for 3 different Iris flowers
plt.plot(setosa["SepalLengthCm"],setosa["SepalWidthCm"],linestyle=" ",marker="o",color="g",markersize=7,label="setosa")
plt.plot(versicolor["SepalLengthCm"],versicolor["SepalWidthCm"],linestyle=" ",marker="o",color="r",markersize=7,label="versicolor")
plt.plot(virginica["SepalLengthCm"],virginica["SepalWidthCm"],linestyle=" ",marker="o",color="k",markersize=7,label="virginica")
plt.xlabel("sepallength")
plt.ylabel("sepalwidth")
plt.title("sl vs sw")
plt.legend()
```

```python
# Plot SepalLengh vs PetalWidth for 3 different Iris flowers
plt.plot(setosa["SepalLengthCm"],setosa["PetalWidthCm"],linestyle=" ",marker="o",color="g",markersize=7,label="setosa")
plt.plot(versicolor["SepalLengthCm"],versicolor["PetalWidthCm"],linestyle=" ",marker="o",color="r",markersize=7,label="versicolor")
plt.plot(virginica["SepalLengthCm"],virginica["PetalWidthCm"],linestyle=" ",marker="o",color="k",markersize=7,label="virginica")
plt.xlabel("Sepallength")
plt.ylabel("Petalwidth")
plt.title("sl vs pw")
plt.legend()
```

```python
# Plot SepalLengh vs PetalLength for 3 different Iris flowers
plt.plot(setosa["SepalLengthCm"],setosa["PetalLengthCm"],linestyle=" ",marker="o",color="g",markersize=7,label="setosa")
plt.plot(versicolor["SepalLengthCm"],versicolor["PetalLengthCm"],linestyle=" ",marker="o",color="r",markersize=7,label="versicolor")
plt.plot(virginica["SepalLengthCm"],virginica["PetalLengthCm"],linestyle=" ",marker="o",color="k",markersize=7,label="virginica")
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLength")
plt.title("sl vs pl")
plt.legend()
```

```python
# Plot SepalWidth vs PetalLength for 3 different Iris flowers
plt.plot(setosa["SepalWidthCm"],setosa["PetalLengthCm"],linestyle=" ",marker="o",color="g",markersize=7,label="setosa")
plt.plot(versicolor["SepalWidthCm"],versicolor["PetalLengthCm"],linestyle=" ",marker="o",color="r",markersize=7,label="versicolor")
plt.plot(virginica["SepalWidthCm"],virginica["PetalLengthCm"],linestyle=" ",marker="o",color="k",markersize=7,label="virginica")
plt.xlabel("SepalWidth")
plt.ylabel("PetalLength")
plt.title("sw vs pl")
plt.legend()
```

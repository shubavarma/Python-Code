```python
import random
p1=0
p2=0
snake_bite={30:7,47:15,56:19,73:51,82:42,92:75,98:55}
ladder={4:25,21:39,29:74,43:76,63:80,71:89}
print("Welcome to Snake and Ladder Game")
player1=input("Name of the First Player: ")
player2=input("Name of the Second Player: ")

def roll_again(pos,player):
    '''this function helps to roll again if the player throws 6 or climbs the ladder'''
    dice=random.randint(1,6)
    if dice==6:
        pos=pos+dice
        print("Number rolled by {}: {}".format(player,dice))
    else:
        pos=pos+dice
        print("Number rolled by {}: {}".format(player,dice))
    return pos

def snakeLadder(pos,player):
    '''this function helps the player throw the dice and advanced based on conditions'''
    dice=random.randint(1,6)
    print("Number rolled by {}: {}".format(player,dice))
    #check if they rolled 6, then need to roll again
    if dice==6 and pos<94:
        pos=roll_again(pos,player)
    
    #check if position is beyond 100
    if pos+dice>100:
        print("Oops number rolled by {} is big, number needed is {}".format(player,(100-pos)))
        return pos
    else:
        pos=pos+dice

    #if player lands at base of ladder
    if pos in ladder:
        print("Heyyyy ladder")
        pos=ladder[pos]
        pos=roll_again(pos,player)
        print("{} is now at: {}".format(player,pos))
    #if player lands at mouth of snake    
    elif pos in snake_bite:
        print("Oops Snake")
        pos=snake_bite[pos]
        print("{} is now at: {}".format(player,pos))
    else:
        print("{} is now at: {}".format(player,pos))
 
    return pos

#loop stops when one of the player lands at 100
while p1<=100 and p2<=100:
    # calling function passing First player parameters
    p1=snakeLadder(p1,player1)
    if p1==100:
        print("Game over {} wins".format(player1))
        break
    # calling function passing Second player parameters
    p2=snakeLadder(p2,player2)
    if p2==100:
        print("Game over {} wins".format(player2))
        break
    
```
